```python
import pandas as pd
from datetime import datetime
%pwd
```

```python
masks_df = pd.read_csv("Public_Mask_Mandates.csv",low_memory=False)
masks_df.drop(columns=['FIPS_State', 'Effective_date', 'Expiration_date','Jurisdiction_Type', 'Issuing_Jurisdiction','FIPS_County','order_code','Effective_NA_Reason','Expiration_NA_Reason', 'Date_Signed', 
                        'origin_dataset','Source_of_Action', 'URL', 'Citation'], inplace = True)
```

```python
masks_df["Current_order_status"].value_counts()
```

```python
date_series = masks_df["date"]
masks_df["Year"]=""
masks_df["Month"]=""
for index, row in masks_df.iterrows():
    masks_df.loc[index,"Year"] = masks_df.loc[index,"date"][6:10]
    masks_df.loc[index,"Month"] = masks_df.loc[index,"date"][:2]
#     if index == 10000:
#         break
```

```python
masks_df.head()
```

```python
masks_df_new = masks_df.groupby(by=["State_Tribe_Territory","Year","Month"])
dff_masks = masks_df_new.count()
Face_Masks_Required_in_Public = dff_masks["Face_Masks_Required_in_Public"]
Opps_to_Wear_Face_Masks_in_Public = dff_masks["date"]
dff_masks["Percent_Public_Face_Mask_Mandate"] = round(100*Face_Masks_Required_in_Public/Opps_to_Wear_Face_Masks_in_Public,2)
```
